```python
import pandas as pd
import requests
```

```python
URL = 'https://stat.leedz.me/api/stat/react/get_raw_grouped_data'

start_date = '2021-12-31'
end_date = '2022-02-16'
token = 'ZuuperUzer123'
```

```python
raw_data = requests.get(f'{URL}?start_date={start_date}&end_date={end_date}&tkn={token}').json()
```

```python
# raw_data
```

```python
columns = ['date', 'week', 'month', 'sub1', 'sub2', 'sub3', 'sub4', 'cost', 'sms_count', 'payout']
data_list = []
for item in raw_data['data']:
    date = item['date']
    week = item['week']
    month = item['month']
    sub1 = item['sub1']
    sub2 = item['sub2']
    sub3 = item['sub3']
    sub4 = item['sub4']
    cost = item['cost']
    sms_count = item['sms_count']
    payout = item['payout']

    data_list.append([date, week, month, sub1, sub2, sub3, sub4, cost, sms_count, payout])

data = pd.DataFrame(columns=columns, data=data_list)
```

```python
data['margin'] = (data['payout'] - data['cost']).astype('int')
data['margin_per_sms'] = round(data['margin'] / data['sms_count'], 2)
```

```python
data.query('sub1 == "sm"').pivot_table(values='margin', aggfunc='sum', columns='sub3', index='date', margins=True)
```
